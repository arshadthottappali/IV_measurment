# PD experiment without UI

This notebook uses the Keithley 2600 TSP backend so the pulse timing runs on the instrument, not in Tkinter.

```python
from pathlib import Path

from pd_notebook_helper import PDConfig, plot_pd_results, run_pd_experiment
```

```python
resource_name = "GPIB0::26::INSTR"
output_path = Path("pd_results.txt")

config = PDConfig(
    pot_v=1.0,
    pot_t=0.01,
    pot_pulses=10,
    read_v=0.1,
    read_t=0.02,
    settle_t=0.01,
    compliance_uA=1.0,
    dep_v=-1.0,
    dep_t=0.01,
    dep_pulses=10,
    gap_delay_s=0.0,
    cycles=1,
    sample_name="sample_01",
    operator="",
    notes="Notebook PD run",
)
```

```python
result = run_pd_experiment(resource_name, config, output_path=str(output_path))
print(result["identity"])
print(f"Read points: {len(result['rows'])}")
print(f"Saved to: {result['output_path']}")
```
